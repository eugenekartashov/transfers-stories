```python
transfers_in = epl92[epl92['transfer_movement'] == 'in']
transfers_in
```

```python
transfers_out = epl92[epl92['transfer_movement'] == 'out']
transfers_out
```

```python
transfers_in_clubs = transfers_in.groupby('club_name', as_index=False)[['fee_cleaned']].sum()
transfers_in_clubs
```

```python
transfers_out_clubs = transfers_out.groupby('club_name', as_index=False)[['fee_cleaned']].sum()
transfers_out_clubs
```

```python
import matplotlib.axes as ax
```

```python
plt.barh(y = 'club_name', width = 'fee_cleaned', align = 'edge', data = transfers_in_clubs)
```

```python
transfers_in_clubs.rename(columns = {'fee_cleaned': 'fee_in'}, inplace = True)
```

```python
transfers_out_clubs.rename(columns = {'fee_cleaned': 'fee_out'}, inplace = True)
```

```python
clubs_transfers = pd.merge(transfers_in_clubs, transfers_out_clubs)
clubs_transfers
```

```python
transfers_in_clubs.rename(columns = {'fee_in': 'fee_cleaned'}, inplace = True)
transfers_in_clubs['inout'] = 'in'
transfers_in_clubs
```

```python
transfers_out_clubs.rename(columns = {'fee_out': 'fee_cleaned'}, inplace = True)
transfers_out_clubs['inout'] = 'out'
transfers_out_clubs
```

```python
club_transfers = pd.concat([transfers_in_clubs, transfers_out_clubs],ignore_index = True,)
```

```python
sns.catplot(y = 'club_name', x = 'fee_cleaned', hue = 'inout', data = club_transfers, kind = 'bar', )
```
